## Background and Prior Work

The National Basketball Association or NBA for short is a professional basketball league which was established in the United States in the year of 1949 after a merger of two rival basketball associations. Since then, the NBA has grown significantly and is a major professional sporting league in the United States. The current NBA consists of three divisions with 30 teams in total split across two conferences and three divisions. Every season of the NBA, there is a championship held where the most successful teams from each conference play against each other (1).

Let us first lay down some basketball definitions which are important for understanding our question and the subsequent work:
* Field-goal: A field goal is a point scored by ordinarily playing the game which can be worth either 2 or 3 points depending on distance.
* Field-goal percentage: This is the percentage of successful field goals made by a team out of the total attempted field goals.
* Three-point shooting percentage: This is the total number of successful three-point field goals made by a team out of the total attempted three-point field goals.
* Free-throw: A free-throw or foul shot is an unopposed shot worth 1 point taken at an area on the basketball court known as the free-throw line and is generally given to players who have been fouled by the opposing team similar to how penalty shots work in other sports.
* Free-throw shooting percentage: This is the number of successful free-throws over the number of total free-throws taken.
* Rebound: This is when the ball is recoverred from a missed field goal or free throw.
* Defensive rebound: This is the number of total rebounds made when in the defending team's half of the basketball court.
* Offensive rebound: This is the number of total rebounds made when in the attacking/opponent's half of the basketball court.
* Assists: This statistic is the number of immediate passes made which led to the a successful field goal.
* Turnovers: These are the number of times a team loses possession of the ball without attempting a field goal.
* Relative Annual Score: There may be a noticeable/significant difference among factors that contribute to victory from year to year.

## Dataset #1: NBA Player Stats by Year

```python
def team_names(short):
    names = {"ATL": "Atlanta Hawks", "BOS": "Boston Celtics", "CHH": "Charlotte Hornets", "CHI": "Chicago Bulls", "CLE": "Cleveland Cavaliers", "DAL": "Dallas Mavericks", "DEN": "Denver Nuggets", "DET": "Detroit Pistons", "GSW": "Golden State Warriors", "HOU": "Houston Rockets", "IND": "Indiana Pacers", "LAC": "Los Angeles Clippers", "LAL": "Los Angeles Lakers", "MIA": "Miami Heat", "MIL": "Milwaukee Bucks", "MIN": "Minnesota Timberwolves", "NJN": "Brooklyn Nets", "NYK": "New York Knicks", "ORL": "Orlando Magic", "PHI": "Philadelphia 76ers", "PHO": "Phoenix Suns", "POR": "Portland Trail Blazers", "SAC": "Sacramento Kings", "SAS": "San Antonio Spurs", "SEA": "Seattle Supersonics", "TOR": "Toronto Raptors", "UTA": "Utah Jazz", "VAN": "Vancouver Grizzlies", "WAS": "Washington Wizards"}
    if(short in names.keys()):
        return names[short]

dfs = {}
for i in range(0, 24):
    if(len(str(i))==1):
        filename = f"./data/basketball-ref200{i}.csv"
        dfs[f"df_200{i}"] = pd.read_csv(filename)
        dfs[f"df_200{i}"] = dfs[f"df_200{i}"].get(["Tm", "FG%", "TOV", "TRB"])
        dfs[f"df_200{i}"] = dfs[f"df_200{i}"].rename(columns={"Tm":"Team", "FG%": "Field Goal %", "TOV": "Turnovers", "TRB": "Total Rebounds"})
        dfs_total_turnovers = dfs[f"df_200{i}"].groupby("Team").sum().get("Turnovers")
        dfs_total_rebounds = dfs[f"df_200{i}"].groupby("Team").sum().get("Total Rebounds")
        dfs[f"df_200{i}"] = dfs[f"df_200{i}"].groupby("Team").mean().get(["Field Goal %"]).assign(Turnovers=dfs_total_turnovers, Total_Rebounds=dfs_total_rebounds).reset_index()
        col_length = dfs[f"df_200{i}"].shape[0]
        years = [int(f"200{i}") for j in range(col_length)]
        dfs[f"df_200{i}"] = dfs[f"df_200{i}"].assign(Year=years)
    else:
        filename = f"./data/basketball-ref20{i}.csv"
        dfs[f"df_20{i}"] = pd.read_csv(filename)
        dfs[f"df_20{i}"] = dfs[f"df_20{i}"].get(["Tm", "FG%", "TOV", "TRB"])
        dfs[f"df_20{i}"] = dfs[f"df_20{i}"].rename(columns={"Tm":"Team", "FG%": "Field Goal %", "TOV": "Turnovers", "TRB": "Total Rebounds"})
        dfs_total_turnovers = dfs[f"df_20{i}"].groupby("Team").sum().get("Turnovers")
        dfs_total_rebounds = dfs[f"df_20{i}"].groupby("Team").sum().get("Total Rebounds")
        dfs[f"df_20{i}"] = dfs[f"df_20{i}"].groupby("Team").mean().get(["Field Goal %"]).assign(Turnovers=dfs_total_turnovers, Total_Rebounds=dfs_total_rebounds).reset_index()
        col_length = dfs[f"df_20{i}"].shape[0]
        years = [int(f"20{i}") for j in range(col_length)]
        dfs[f"df_20{i}"] = dfs[f"df_20{i}"].assign(Year=years)
all_years = pd.concat(list(dfs.values()), ignore_index=True).sort_values(by="Year", ascending=True)
all_years["Team"] = all_years["Team"].apply(team_names)
all_years = all_years[all_years['Team'] != 'TOT']
all_years = all_years.groupby(["Team", "Year"]).max()
all_years
```

## Part 2: Distribution of Field Goal % overtime

Well that is immediately not good for our hypothesis but maybe this year was an abnormality, lets check the field goal % across all years and their champion's field goal %.
The graphs will be labeled with their year + the champions field goal % that year. This will allow us to figure out the range of field goal % across all years and the relation it has to the champions stats. 

* Note the last graph is empty, this is simply due to the format I used along with the number of years we had to work with. It doesn't mean anything

```python
def plot_metric_distribution(metric, title, xlabel, ylabel, champions, all_years):

    fig, axes = plt.subplots(5, 5, figsize=(20, 25), sharex=False, sharey=False)
    axes = axes.flatten()
    fig.suptitle(f'{title} over the years', fontsize=20)

    for i in range(24):
        currYear = str(i + 2000)
        year_data = all_years.filter(like=currYear, axis=0)
        axes[i].hist(year_data[metric], bins=5, edgecolor='black')

        champion = champions[champions['Year'] == i + 2000]['NBA Champion'].values
        metric_val = year_data.loc[(champion[0], i + 2000)][metric]
        metric_val *= 100
        metric_val = int(metric_val)
        metric_val /= 100

        axes[i].set_title(f'{currYear} : Champion {metric} {metric_val}%')
        axes[i].set_xlabel(f'{xlabel}')
        axes[i].set_ylabel(f'{ylabel}')

    plt.show()

plot_metric_distribution('Field Goal %', 'Field Goal Percentage Histograms', 'Field Goal %', '# of teams', champions, all_years)
```

Looking at all these graphs you might see that amoung their years the champions almost never get the most field goal %, in fact they don't even get close generally staying in the middle of the pack with only a few outliers like 2015 or 2006.
